```python
# 1. Import các thư viện cần thiết
import numpy as np
from scipy import linalg, sparse

# 2. Tạo và in ma trận D
D = np.mat([[3, 4], [5, 6]])
print("Ma trận D:")
print(D)

# 3. Tạo ma trận C với kích thước 5x7 gồm các số ngẫu nhiên
C = np.mat(np.random.random((5, 7)))
print("\nMa trận C:")
print(C)

# 4. Tạo ma trận vuông A 2x2 với số ngẫu nhiên
A = np.mat(np.random.random((2, 2)))
print("\nMa trận A:")
print(A)

# 5. Tạo mảng b chứa số thực và số phức, sau đó chuyển thành ma trận B
b = np.array([[(1 + 5j), 2j, 3j], (4, 5, 6)])
B = np.asmatrix(b)
print("\nMa trận B (chuyển từ mảng có số phức):")
print(B)
# Tạo một ma trận A vuông để minh họa
A = np.array([[2, 1], [5, 3]])

# 6. Chuyển vị và nghịch đảo
print("Ma trận A:")
print(A)

print("\nChuyển vị của A (A.T):")
print(A.T)

print("\nMa trận nghịch đảo của A (A.I hoặc linalg.inv(A)):")
print(np.linalg.inv(A))  # hoặc A.I nếu dùng np.mat

# 8. Ma trận dưới từ đường chéo
M = np.array([[-1, 3, 2],
              [ 0, -2, 1],
              [ 1, 5, -2]])
M_lower = np.tril(M)
print("\nMa trận M:")
print(M)

print("\nMa trận dưới từ đường chéo (np.tril(M)):")
print(M_lower)

# 9. Ma trận trên từ đường chéo
M_upper = np.triu(M)
print("\nMa trận trên từ đường chéo (np.triu(M)):")
print(M_upper)

# 10. Vector và ma trận đường chéo
v_diag = np.diag(M)  # Lấy vector đường chéo
print("\nVector đường chéo (np.diag(M)):")
print(v_diag)

M_diag = np.diag(v_diag)  # Tạo lại ma trận đường chéo từ vector
print("\nMa trận đường chéo từ vector (np.diag(v_diag)):")
print(M_diag)
# Tạo ma trận A 6x6 gồm các số từ 0 đến 35
A = np.reshape(np.arange(36.0), (6, 6))
print("Ma trận A:")
print(A)

# Tạo ma trận đơn vị I6
I6 = np.identity(6)
print("\nMa trận đơn vị I6:")
print(I6)

# Kích thước của A (số phần tử)
print("\nSố phần tử của ma trận A (A.size):", A.size)

# Lấy đường chéo chính của ma trận A
diagonal_A = np.matrix.diagonal(A)
print("\nĐường chéo chính của A:")
print(diagonal_A)

# Tạo ma trận A mới: A = A + I
A = A + I6
print("\nMa trận A sau khi cộng với I (A + I):")
print(A)

# Tính tích của A với 1 vector (ví dụ vector hàng gồm toàn 1)
v = np.ones((6, 1))
product = A @ v
print("\nTích A nhân với vector v:")
print(product)
```

```python
# Tạo lại ma trận A 6x6
A = np.reshape(np.arange(36.0), (6, 6))
print("Ma trận A:")
print(A)

# Tạo vector vecB và nhân với A
vecB = np.array([1., 2., 3., 4., 5., 6.])
C = A.dot(vecB)
print("\nVector vecB:")
print(vecB)
print("\nKết quả C = A.dot(vecB):")
print(C)

# Tạo ma trận D (2x6) và nhân với A
D = np.array([[1., 2., 3., 4., 5., 6.],
              [1., 0., 1., 0., 1., 0.]])
print("\nMa trận D:")
print(D)

E = A.dot(D.T)  # Transpose vì A là (6x6), D.T là (6x2) -> kết quả (6x2)
print("\nKết quả E = A.dot(D.T):")
print(E)

# Tạo ma trận F (6x2)
F = np.array([[1., 1.],
              [2., 0.],
              [3., 1.],
              [4., 0.],
              [5., 1.],
              [6., 0.]])
print("\nMa trận F:")
print(F)

G = A.dot(F)
print("\nKết quả G = A.dot(F):")
print(G)
# 1. Tính tích A × vecB
vecB = np.array([1., 2., 3., 4., 5., 6.])
C = A.dot(vecB)
print("\nVector vecB:")
print(vecB)
print("\nKết quả C = A.dot(vecB):")
print(C)

# 2. Khai báo ma trận D 2x6
D = np.array([[1., 2., 3., 4., 5., 6.],
              [1., 0., 1., 0., 1., 0.]])
print("\nMa trận D:")
print(D)

# 3. Tính tích A × D.T (chuyển vị vì A là 6x6, D là 2x6)
E = A.dot(D.T)
print("\nKết quả E = A.dot(D.T):")
print(E)

# 4. Xây dựng ma trận F kích thước 6x2
F = np.array([[1., 1.],
              [2., 0.],
              [3., 1.],
              [4., 0.],
              [5., 1.],
              [6., 0.]])
print("\nMa trận F:")
print(F)

# 5. Tính tích A × F
G = A.dot(F)
print("\nKết quả G = A.dot(F):")
print(G)
# Ma trận F kích thước 6x2
F = np.array([[1., 1.],
              [2., 0.],
              [3., 1.],
              [4., 0.],
              [5., 1.],
              [6., 0.]])
print("\nMa trận F:")
print(F)

# Tính G = A × F
G = A.dot(F)
print("\nMa trận G = A.dot(F):")
print(G)

# Tính ma trận nghịch đảo của A (nếu khả thi)
try:
    A_inv = np.linalg.inv(A)
    print("\nMa trận nghịch đảo của A:")
    print(A_inv)
except np.linalg.LinAlgError:
    print("\nMa trận A không khả nghịch (không thể lấy nghịch đảo).")

# Tính np.linalg.inv(np.linalg.inv(A))
try:
    A_double_inv = np.linalg.inv(np.linalg.inv(A))
    print("\nKết quả của np.linalg.inv(np.linalg.inv(A)):")
    print(A_double_inv)
except np.linalg.LinAlgError:
    print("\nKhông thể tính np.linalg.inv(np.linalg.inv(A)).")
```

```python
import sympy as sp
from sympy import Symbol, solve

# Khai báo các ẩn số
x1 = Symbol('x1')
x2 = Symbol('x2')
x3 = Symbol('x3')
x4 = Symbol('x4')

# Khai báo các phương trình
pt1 = x4 + 610 - 450 - x1        # tương đương: x4 - x1 + 160 = 0
pt2 = x1 + 400 - x2 - 640        # tương đương: x1 - x2 - 240 = 0
pt3 = x2 + 600 - x3              # tương đương: x2 - x3 + 600 = 0
pt4 = x3 - x4 - 520              # tương đương: x3 - x4 - 520 = 0

# Giải hệ phương trình
nghiem = sp.solve((pt1, pt2, pt3, pt4))

# In nghiệm
print(nghiem)
```
